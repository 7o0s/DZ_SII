```python
def parse_review_page(URL:str):
    """
    Gets information about review about game on a page (name, name of article, text of review) and
    returns it as dict.
    :param URL: str - page url
    :return: None or dict full of info (keys: name, name_review, ref, text)
    """
    r = requests.get(URL, headers=headers, timeout=600)
    soup = bs(r.text, "html.parser")
    name = soup.find(class_="post-title entry-title")
    game_name = name.text.partition('review')[0]
    waste_block = soup.find(class_="entry-content clearfix").select('div.appip-block-wrapper')
    for w in waste_block:
        w.extract()
    game_review = soup.find(class_="entry-content clearfix").select('p')
    game_full_review = ""
    for review in game_review[:-2]:
        game_full_review += review.text
    waste_date = soup.find(class_ = "published").text.partition(',')
    date = waste_date[0][:-2] + waste_date[1] + waste_date[2]
    if soup.find(class_ = "review_grade") is not None:
        if soup.find(class_ = "review_grade").span is not None:
            grade = soup.find(class_ = "review_grade").span.text
        else: 
            grade = soup.find(class_ = "review_grade").text.partition('Grade: ')[2]
    else:
        grade = None
    d = {'name': game_name, 'ref': URL, 'text': game_full_review, 'date': date, 'grade': grade}
    return d

filename = '/Users/karina/Desktop/НИУ ВШЭ/3 курс/ПроектГеймДев/HardcoreGamer/game_URL.txt'
counter = 0
with open (filename, 'r') as f:
    all_review_links = f.readlines()
for link in all_review_links[counter:]:
    d = parse_review_page("%s" %link.rstrip('\n'))
    with open("/Users/karina/Desktop/НИУ ВШЭ/3 курс/ПроектГеймДев/HardcoreGamer/Reviews/game_review_all_" + str(counter) + ".json", "w") as outfile:
        outfile.write(json.dumps(d))
    counter = counter + 1
```

```python
def parse_review_page(URL:str):
    """
    Gets information about review about game on a page (name, name of article, text of review) and
    returns it as dict.
    :param URL: str - page url
    :return: None or dict full of info (keys: name, name_review, ref, text)
    """
    r = requests.get(URL, headers=headers, timeout=600)
    soup = bs(r.text, "html.parser")
    name = soup.find(class_="entry-title").text.partition('Review:')[2]
    game_review = soup.find(class_="entry-content")
    date = soup.find(class_="entry-meta").text.partition('on ')[2]
    links = game_review.find_all('div', class_="score-footer-wrapper")
    for link in links:
        grade = int(str(link.find('div')).partition('score s')[2].partition('"></div>')[0])/10
    links = game_review.find_all('p')
    game_full_review = ''
    for link in links[:-1]:
        game_full_review += link.text
    d = {'name': name, 'ref': URL, 'text': game_full_review, 'date': date, 'grade': grade}
    return d
def parse_review(URL:str):
    r = requests.get(URL, headers=headers, timeout=600)
    soup = bs(r.text, "html.parser")
    links = soup.find('div', class_="score-footer-wrapper")
    print(links.find('p'))
    if links.find('p') != None:
        return True
    else:
        return False
    

filename = '/Users/karina/Desktop/НИУ ВШЭ/3 курс/ПроектГеймДев/HardcoreGamer/game_URL.txt'
counter = 2273
counter2 = 3810
i = 2021
with open (filename, 'r') as f:
    all_review_links = f.readlines()
for link in all_review_links[counter:counter2]:
    flag = parse_review("%s" %link.rstrip('\n'))
    print(flag)
    if flag:
        d = parse_review_page("%s" %link.rstrip('\n'))
        with open("/Users/karina/Desktop/НИУ ВШЭ/3 курс/ПроектГеймДев/HardcoreGamer/Reviews/game_review_all_" + str(i) + ".json", "w") as outfile:
            outfile.write(json.dumps(d))
        i += 1
    counter = counter + 1
```
